```python
from statsmodels.tsa.stattools import adfuller 
def stationarity_check(ts):
    
    # Determing rolling statistics
    #roll_mean = pd.rolling_mean(ts, window=12)
    roll_mean = ts.rolling(12).mean()
    # Plot rolling statistics:
    plt.plot(ts, color='green',label='Original')
    plt.plot(roll_mean, color='blue', label='Rolling Mean')
    plt.legend(loc='best')
    plt.title('Rolling Mean')
    plt.show(block=False)
    
    # Perform Augmented Dickey-Fuller test:
    print('Augmented Dickey-Fuller test:')
    df_test = adfuller(ts)
    print("type of df_test: ",type(df_test))
    print("df_test: ",df_test)
    df_output = pd.Series(df_test[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    print("df_output: \n",df_output)
    for key,value in df_test[4].items():
        df_output['Critical Value (%s)'%key] = value
    print(df_output)
```

```python
stationarity_check(df_Colombo.AverageTemperature)
```

```python
from statsmodels.graphics.tsaplots import plot_pacf,plot_acf
```

```python
plot_acf(df_Colombo.AverageTemperature, lags=50)
plot_pacf(df_Colombo.AverageTemperature, lags=50)
plt.xlabel('lags')
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARMA
```

```python
import itertools
p = q = range(0, 4)
pq = itertools.product(p, q)
for param in pq:
    try:
        mod = ARMA(df_Colombo.AverageTemperature,order=param)
        results = mod.fit()
        print('ARMA{} - AIC:{}'.format(param, results.aic))
    except:
        continue
```

```python
model = ARMA(df_Colombo.AverageTemperature, order=(2,3))  
results_MA = model.fit()  
plt.plot(df_Colombo.AverageTemperature)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('Fitting data _ MSE: %.2f'% (((results_MA.fittedvalues-df_Colombo.AverageTemperature)**2).mean()))
plt.show()
```

```python
predictions = results_MA.predict('01/01/1796', '09/01/2014')
```
